```python
def create_test_project_structure(base_path: str = "/tmp/test_agent_project") -> str:
    """Создает тестовую структуру проекта для безопасного тестирования."""
    import shutil
    from pathlib import Path
    
    # Очищаем предыдущую структуру
    if os.path.exists(base_path):
        shutil.rmtree(base_path)
    
    base = Path(base_path)
    base.mkdir(parents=True, exist_ok=True)
    
    # Создаем файлы
    (base / "main.py").write_text("""#!/usr/bin/env python3
\"\"\"Main application entry point.\"\"\"
def main():
    print("Hello from main!")
    return 0

if __name__ == "__main__":
    main()
""")
    
    (base / "utils.py").write_text("""\"\"\"Utility functions.\"\"\"
def helper(data: str) -> str:
    \"\"\"Process input data.\"\"\"
    return data.upper()

def validate_input(value: Any) -> bool:
    \"\"\"Validate input value.\"\"\"
    return value is not None
""")
    
    (base / "config.json").write_text("""{
    "debug": true,
    "timeout": 30,
    "features": {
        "enable_cache": true,
        "log_level": "INFO"
    }
}""")
    
    # Создаем поддиректории
    (base / "services").mkdir(exist_ok=True)
    (base / "services" / "data_service.py").write_text("""\"\"\"Data service implementation.\"\"\"
class DataService:
    def __init__(self, connection_string: str):
        self.connection_string = connection_string
    
    def fetch_data(self, query: str):
        return f"Executed: {query}"
""")
    
    (base / "tests").mkdir(exist_ok=True)
    (base / "tests" / "test_utils.py").write_text("""import pytest
from utils import helper

def test_helper():
    assert helper("test") == "TEST"
""")
    
    logger.info(f"✓ Тестовая структура создана: {base}")
    logger.info(f"  Файлы: {[f.name for f in base.rglob('*') if f.is_file()]}")
    return str(base)

TEST_PROJECT_PATH = create_test_project_structure()
```

```python
async def initialize_test_system_context() -> SystemContext:
    """Инициализирует системный контекст с моками для безопасного тестирования."""
    # Создаем мок-провайдер LLM
    from unittest.mock import AsyncMock
    from domain.models.llm.llm_types import LLMResponse, LLMRequest
    
    class MockLLMProvider:
        def __init__(self):
            self.name = "mock_llm"
            self.is_initialized = False
            
        async def initialize(self) -> bool:
            self.is_initialized = True
            logger.info("✓ Mock LLM провайдер инициализирован")
            return True
        
        async def generate(self, request: LLMRequest) -> LLMResponse:
            # Простая эмуляция ответа на основе промпта
            if "анализ" in request.prompt.lower() or "structure" in request.prompt.lower():
                content = """{
                    "project_summary": "Тестовый проект с 4 файлами",
                    "main_components": ["main.py", "utils.py", "services/", "tests/"],
                    "potential_issues": [],
                    "recommendations": ["Добавить документацию к функциям"]
                }"""
            elif "навигация" in request.prompt.lower() or "navigate" in request.prompt.lower():
                content = """{
                    "found": true,
                    "file_path": "utils.py",
                    "line_number": 4,
                    "signature": "def helper(data: str) -> str",
                    "context": "Utility function for string processing"
                }"""
            else:
                content = f"Mock response to: {request.prompt[:50]}..."
            
            return LLMResponse(
                content=content,
                model="mock-model",
                tokens_used=len(request.prompt) // 4,
                generation_time=0.05,
                finish_reason="stop"
            )
    
    # Создаем мок-провайдер БД
    class MockDBProvider:
        def __init__(self):
            self.name = "mock_db"
            self.is_initialized = False
            
        async def initialize(self) -> bool:
            self.is_initialized = True
            logger.info("✓ Mock DB провайдер инициализирован")
            return True
        
        async def execute(self, query: str, params: Dict = None) -> Dict[str, Any]:
            # Эмуляция ответа БД
            if "SELECT" in query.upper():
                return {
                    "success": True,
                    "rows": [
                        {"id": 1, "name": "test_item", "created_at": datetime.now().isoformat()}
                    ],
                    "rowcount": 1,
                    "columns": ["id", "name", "created_at"],
                    "execution_time": 0.01
                }
            return {"success": True, "rowcount": 0, "execution_time": 0.005}
    
    # Инициализация системного контекста
    context = SystemContext()
    
    # Регистрируем моки вручную (для тестов)
    mock_llm = MockLLMProvider()
    mock_db = MockDBProvider()
    
    await mock_llm.initialize()
    await mock_db.initialize()
    
    # Сохраняем моки в атрибуты контекста для доступа из тестов
    context._mock_llm = mock_llm
    context._mock_db = mock_db
    
    # Инициализируем контекст
    await context.initialize()
    
    logger.info(f"✓ SystemContext инициализирован")
    logger.info(f"  LLM провайдер: {mock_llm.name} (initialized={mock_llm.is_initialized})")
    logger.info(f"  DB провайдер: {mock_db.name} (initialized={mock_db.is_initialized})")
    
    return context

# Запуск инициализации
system_context = await initialize_test_system_context()
```

```python
async def test_file_tools(context: SystemContext, project_path: str):
    """Тестирует работу файловых инструментов."""
    logger.info("\n" + "=" * 80)
    logger.info("ТЕСТИРОВАНИЕ ИНСТРУМЕНТОВ")
    logger.info("=" * 80)
    
    # Получаем инструменты из контекста
    file_lister = context.get_resource("file_lister")
    file_reader = context.get_resource("file_reader")
    
    if not file_lister or not file_reader:
        logger.error("✗ Инструменты не найдены в контексте!")
        return
    
    logger.info("\n📁 Тест 1: Список файлов (FileListerTool)")
    try:
        result = await file_lister.execute({"path": project_path, "recursive": True})
        logger.info(f"  ✓ Успешно получено {len(result.get('files', []))} файлов")
        for f in result.get('files', [])[:3]:  # Первые 3 файла
            logger.info(f"    - {f}")
        if len(result.get('files', [])) > 3:
            logger.info(f"    ... и ещё {len(result.get('files', [])) - 3} файлов")
    except Exception as e:
        logger.error(f"  ✗ Ошибка: {e}")
    
    logger.info("\n📄 Тест 2: Чтение файла (FileReaderTool)")
    try:
        test_file = os.path.join(project_path, "utils.py")
        result = await file_reader.execute({"path": test_file})
        content_preview = result.get('content', '')[:100].replace('\n', ' ')
        logger.info(f"  ✓ Прочитано {len(result.get('content', ''))} байт")
        logger.info(f"  💡 Превью: {content_preview}...")
    except Exception as e:
        logger.error(f"  ✗ Ошибка: {e}")
    
    logger.info("\n🔒 Тест 3: Защита от опасных путей")
    try:
        dangerous_path = os.path.join(project_path, "../etc/passwd")
        result = await file_reader.execute({"path": dangerous_path})
        logger.warning(f"  ⚠ Инструмент не заблокировал опасный путь!")
    except Exception as e:
        logger.info(f"  ✓ Опасный путь успешно заблокирован: {type(e).__name__}")
    
    logger.info("\n" + "=" * 80)

await test_file_tools(system_context, TEST_PROJECT_PATH)
```

```python
async def test_llm_calls(context: SystemContext):
    """Тестирует вызовы LLM через системный контекст."""
    logger.info("\n" + "=" * 80)
    logger.info("ТЕСТИРОВАНИЕ LLM")
    logger.info("=" * 80)
    
    # Тест 1: Прямой вызов через мок-провайдер
    logger.info("\n💬 Тест 1: Прямой вызов мок-провайдера")
    try:
        request = type('LLMRequest', (), {
            'prompt': 'Кратко опиши структуру тестового проекта',
            'max_tokens': 150,
            'temperature': 0.3
        })
        response = await context._mock_llm.generate(request)
        logger.info(f"  ✓ Получен ответ ({len(response.content)} символов)")
        # Показываем первые 200 символов ответа
        preview = response.content[:200].replace('\n', ' ')
        logger.info(f"  💬 Превью: {preview}...")
    except Exception as e:
        logger.error(f"  ✗ Ошибка: {e}")
    
    # Тест 2: Вызов через интерфейс контекста (если реализован)
    logger.info("\n🤖 Тест 2: Вызов через интерфейс контекста")
    try:
        # Если в вашей архитектуре есть такой метод:
        if hasattr(context, 'call_llm'):
            response = await context.call_llm(
                user_prompt="Назови 3 основных файла в проекте",
                system_prompt="Ты — эксперт по анализу кода",
                max_tokens=100
            )
            logger.info(f"  ✓ Ответ получен: {response.content[:150]}...")
        else:
            logger.info("  ℹ Метод call_llm не реализован в текущей версии")
    except Exception as e:
        logger.warning(f"  ⚠ Не поддерживается: {e}")
    
    logger.info("\n" + "=" * 80)

await test_llm_calls(system_context)
```

```python
async def test_skills(context: SystemContext, project_path: str):
    """Тестирует работу навыков."""
    logger.info("\n" + "=" * 80)
    logger.info("ТЕСТИРОВАНИЕ НАВЫКОВ")
    logger.info("=" * 80)
    
    # Тест 1: Навык анализа проекта
    logger.info("\n🗺 Тест 1: ProjectMapSkill (анализ структуры проекта)")
    try:
        project_map_skill = context.get_resource("project_map")
        if project_map_skill:
            # Подготавливаем входные данные
            input_data = {
                "root_path": project_path,
                "include_patterns": ["*.py", "*.json"],
                "exclude_patterns": ["__pycache__", ".git", "tests/*"],
                "max_depth": 3,
                "detail_level": "signature"
            }
            
            # Создаем сессию для навыка
            session = SessionContext()
            await session.initialize()
            
            # Выполняем навык
            result = await project_map_skill.execute(
                capability=None,  # Для тестов можно передать None
                parameters=input_data,
                context=session
            )
            
            if result and hasattr(result, 'success') and result.success:
                logger.info(f"  ✓ Анализ завершён успешно")
                # Показываем структуру результата
                if hasattr(result, 'result') and result.result:
                    logger.info(f"  📊 Найдено файлов: {len(result.result.get('files', []))}")
                    logger.info(f"  📂 Найдено директорий: {len(result.result.get('directories', []))}")
            else:
                logger.warning(f"  ⚠ Навык вернул результат без флага success")
        else:
            logger.info("  ℹ ProjectMapSkill не зарегистрирован в контексте")
    except Exception as e:
        logger.error(f"  ✗ Ошибка: {e}")
        import traceback
        logger.debug(traceback.format_exc())
    
    # Тест 2: Навык навигации (если доступен)
    logger.info("\n🧭 Тест 2: ProjectNavigatorSkill (навигация по коду)")
    try:
        navigator = context.get_resource("project_navigator")
        if navigator:
            session = SessionContext()
            await session.initialize()
            
            # Устанавливаем структуру проекта в контекст сессии
            from domain.models.project_structure import ProjectStructure
            project_structure = ProjectStructure(
                name="test_project",
                root_path=project_path,
                files=["main.py", "utils.py", "config.json", "services/data_service.py"],
                directories=["services", "tests"]
            )
            session.set_context({"project_structure": project_structure})
            
            # Выполняем поиск
            result = await navigator.execute(
                capability=None,
                parameters={
                    "query": "helper",
                    "search_type": "function",
                    "file_path": os.path.join(project_path, "utils.py")
                },
                context=session
            )
            
            if result and result.success:
                logger.info(f"  ✓ Функция найдена: {result.result.get('identifier', 'N/A')}")
                logger.info(f"  📍 Расположение: {result.result.get('file_path', 'N/A')}")
            else:
                logger.info(f"  ℹ Функция не найдена или навык вернул ошибку")
        else:
            logger.info("  ℹ ProjectNavigatorSkill не зарегистрирован")
    except Exception as e:
        logger.error(f"  ✗ Ошибка: {e}")
    
    logger.info("\n" + "=" * 80)

await test_skills(system_context, TEST_PROJECT_PATH)
```

```python
async def integration_test_agent():
    """Полный интеграционный тест агента."""
    logger.info("\n" + "=" * 80)
    logger.info("ИНТЕГРАЦИОННЫЙ ТЕСТ АГЕНТА")
    logger.info("=" * 80)
    
    # 1. Создаем сессию
    logger.info("\n1️⃣  Создание сессии")
    session = SessionContext()
    await session.initialize()
    session.set_goal("Проанализировать структуру тестового проекта и найти функцию 'helper'")
    logger.info(f"  ✓ Сессия создана, цель установлена")
    
    # 2. Создаем агента
    logger.info("\n2️⃣  Инициализация агента")
    agent = AgentRuntime(
        system_context=system_context,
        session_context=session
    )
    logger.info(f"  ✓ Агент инициализирован")
    
    # 3. Адаптация к задаче
    logger.info("\n3️⃣  Адаптация к задаче")
    adaptation = agent.adapt_to_task(session.get_goal())
    logger.info(f"  🎯 Определён домен: {adaptation.get('domain', 'N/A')}")
    logger.info(f"  🧠 Рекомендованный паттерн: {adaptation.get('pattern', 'N/A')}")
    
    # 4. Выполнение шага (упрощённо)
    logger.info("\n4️⃣  Выполнение первого шага")
    try:
        # Получаем навык анализа проекта
        project_map_skill = system_context.get_resource("project_map")
        if project_map_skill:
            result = await project_map_skill.execute(
                capability=None,
                parameters={"root_path": TEST_PROJECT_PATH},
                context=session
            )
            if result and result.success:
                logger.info(f"  ✓ Шаг выполнен успешно")
                # Сохраняем результат в контекст
                session.set_context({"project_analysis": result.result})
            else:
                logger.warning(f"  ⚠ Шаг завершился с предупреждением")
        else:
            logger.info(f"  ℹ ProjectMapSkill недоступен для выполнения")
    except Exception as e:
        logger.error(f"  ✗ Ошибка выполнения: {e}")
    
    # 5. Проверка состояния
    logger.info("\n5️⃣  Проверка состояния агента")
    state = agent.state if hasattr(agent, 'state') else AgentState()
    logger.info(f"  📊 Шаг: {state.current_step}")
    logger.info(f"  ✅ Ошибок: {state.error_count}")
    logger.info(f"  📈 Прогресс: {state.no_progress_steps} шагов без прогресса")
    
    # 6. Финальный отчёт
    logger.info("\n6️⃣  Финальный отчёт")
    context_data = session.get_context()
    if "project_analysis" in context_data:
        analysis = context_data["project_analysis"]
        files = analysis.get("files", [])
        logger.info(f"  📁 Проанализировано файлов: {len(files)}")
        if files:
            logger.info(f"  📄 Примеры: {', '.join([os.path.basename(f) for f in files[:3]])}")
    else:
        logger.info(f"  ℹ Данные анализа отсутствуют в контексте")
    
    # 7. Очистка
    logger.info("\n7️⃣  Очистка ресурсов")
    await session.shutdown()
    logger.info(f"  ✓ Сессия завершена")
    
    logger.info("\n" + "=" * 80)
    logger.info("✅ ИНТЕГРАЦИОННЫЙ ТЕСТ ЗАВЕРШЁН")
    logger.info("=" * 80)

await integration_test_agent()
```

```python
from tree_sitter import Tree
from tree_sitter import Language, Parser
import tree_sitter_python

# 1. Инициализация парсера (например, для Python)
PY_LANGUAGE = Language(tree_sitter_python.language())
parser = Parser(language = PY_LANGUAGE)

# 2. Исходный код
source_code = b"def hello(): print('world')"

# 3. Парсинг
tree = parser.parse(source_code)
root_node = tree.root_node

# --- Альтернатива: рекурсивная печать ---
def print_tree(node, indent=0):
    print(' ' * indent + f"{node.type} [{node.start_point} - {node.end_point}]")
    for child in node.children:
        print_tree(child, indent + 2)

print_tree(root_node)
```
